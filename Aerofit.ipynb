<h1>Graphical Analysis with Inferences</h1>

````python
labels = ['KP281 (Entry-Level)', 'KP481 (Mid-Level)', 'KP781 (Advanced-Level)']
label = ['KP281', 'KP481', 'KP781']
units_sold = [len(entryData), len(midData), len(advData)] 

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(13, 7)) 
axes[0].bar(label, units_sold)

bars = axes[0].bar(label, units_sold)

for bar in bars:
    height = bar.get_height()
    axes[0].text(bar.get_x() + bar.get_width()/2, 1.01*height, f"{height}",ha='center', va='bottom')


axes[0].set_xlabel('Treadmill Level')
axes[0].set_ylabel('Units Sold')
axes[0].set_title('Treadmill Sales Distribution (Total Sales By Count)')

axes[1].pie(units_sold, labels=labels, autopct="%1.2f%%", wedgeprops=dict(edgecolor='black', linewidth=1))  
axes[1].set_title('Treadmill Sales Distribution (Percentage By Count)')

plt.tight_layout()
plt.show()
````

The above graphs show the distrubtion of sales by count of sales across all the product

It can be seen that the ```KP281``` is sold almost twice as much as the ```KP781``` 

- Entry-Level treadmills are the most popular category based on unit sales.
- Mid-Level treadmills hold the second position in sales metrics.
- Advanced-Level treadmills have the lowest sales volume.

````python
labels = ['KP281 (Entry-Level)', 'KP481 (Mid-Level)', 'KP781 (Advanced-Level)']
label = ['KP281', 'KP481', 'KP781']
total_sales_price = [len(entryData)*1500, len(midData)*1750, len(advData)*2500] 

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(13, 7)) 
axes[0].bar(label, units_sold)


bars = axes[0].bar(label, total_sales_price)

for bar in bars:
    height = bar.get_height()
    axes[0].text(bar.get_x() + bar.get_width()/2., 1.02*height, f"${height}",ha='center', va='bottom')

axes[0].set_xlabel('Treadmill Level')
axes[0].set_ylabel('Revenue Generated ($)')
axes[0].set_title('Treadmill Sales Distribution (Total Sales By Value)')

axes[1].pie(total_sales_price, labels=labels, autopct="%1.2f%%", wedgeprops=dict(edgecolor='black', linewidth=1))  
axes[1].set_title('Treadmill Sales Distribution (Percentage By Value)')

plt.tight_layout()
plt.show()
````

The above graphs show the distrubtion of sales by count of sales across all the product

It can be seen that even though ```KP281``` was sold almost twice as much as the ```KP781``` the sales dont differ that much this is because of the higher average order value of the ```KP781``` Treadmill 

- Entry-Level treadmills are the biggest revenue generator for Aerofit but not by a huge margin generating around ```$120000``` in revenue.
- Mid-Level treadmills hold the second position in sales by value generating ```$105000$``` in revenue.
- Advanced-Level treadmills have the lowest revenue at ```$100000``` but they are very close to Mid-Level Treadmills laggng behind by just ```$5000```.

````python
product_counts = df[prod].value_counts(normalize=1)

total_count = product_counts.sum()
product_probabilities = product_counts / total_count

labels = ["KP281 (Entry-Level)","KP481 (Mid-Level)","KP781(Advanced-Level)"]

heatmap_data = product_probabilities.values.reshape(-1, 1).T

plt.figure(figsize=(10, 2))
sb.heatmap(heatmap_data, annot=True, fmt=".3f", cmap='flare', xticklabels=labels, yticklabels=['']) 
plt.title('Product Purchase Probability Heatmap')
plt.show()
````

The most probable buy is Entry Level (P = 0.444)

# Recommendations for Boosting Sales


### Strategies Targeting All Customer Segments

* **Emphasize Health Benefits:** Messaging should focus on cardiovascular health, stress reduction, and improved energy levels, beyond just weight loss.
* **"Try Before You Buy" Options:** Offer in-store trials or short-term rentals to let customers experience the treadmills.
* **Flexible Financing:**  Easy financing plans with low monthly payments increase accessibility.
* **Highlight Smart Features:** Promote built-in workout programs, connectivity, and entertainment options (especially for the KP781).

### Strategies Focused on KP281

* **Target Young Professionals:** Position treadmills as convenient, time-saving home-fitness solutions.
* **"Starter Package" Bundles:** Offer the KP281 with accessories (mats, heart rate monitors) at a slight discount.
* **Partnerships with Fitness Influencers:** Collaborate with influencers popular in the 15-35 age range.

### Strategies Focused on KP481

* **Couples Promotions:** Create promotions around Valentine's Day, anniversaries, etc., with "Get fit together" messaging.
* **"Family Fitness" Emphasis:** Position treadmills for whole-family health improvement.

### Strategies Focused on KP781

* **Content for Fitness Enthusiasts:** Create content demonstrating advanced training plans possible with the KP781.
* **Partnerships with Gyms/Trainers:** Offer the KP781 to small gyms/trainers with potential referral programs.
* **Justify the Upgrade:**
     Clearly outline the enhanced features of the KP781:-
    * **Motor Power:** Explain how the more powerful motor supports high-intensity workouts and accommodates heavier users.
